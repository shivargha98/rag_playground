```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
embedding_model = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
text = "India beat austrailia in semi final of the women's worldcup"
embedding = embedding_model.embed_query(text)
print(f"lenght of embedding {len(embedding)}")
print(f"Embeddings:{embedding}")
```

```python
##semantic search example###
documents = ["The Amazon rainforest produces about 20% of the world's oxygen and is home to millions of species.",
"Machine learning models improve automatically through experience without being explicitly programmed.",
"Meditation helps reduce anxiety by training the mind to focus on the present moment.",
"Electric vehicles are becoming more popular as battery technology improves and charging stations expand.",
"Regular exercise strengthens the heart, increases endurance, and boosts overall mood."]
```

```python
import numpy as np
def cosine_similarity(vect1,vect2):
    '''calculates the cosine similarity between 2 vectors
    formula : (A.B)/ (||A||*||B||)'''
    dot_prod = np.dot(vect1,vect2) ##dot product
    norm_vect1 = np.linalg.norm(vect1) ## normalisation of vector-1
    norm_vect2 = np.linalg.norm(vect2) ## normalisation of vector-2

    return (dot_prod)/(norm_vect1 * norm_vect2)

def semantic_search(query,documents,top_k):

    similarity_score_list = []

    query_embedding = embedding_model.embed_query(query)
    doc_embeddings = embedding_model.embed_documents(documents) ##all docs as embeddings##

    for index,doc_embed in enumerate(doc_embeddings):
        similarity_score = cosine_similarity(query_embedding,doc_embed)
        similarity_score_list.append([similarity_score,documents[index]])
    
    similarity_score_list.sort(reverse=True)
    return similarity_score_list[:top_k]
```

```python
query = 'What is Machine Learning?'
print(f"Input query : {query}")
semantic_search_res = semantic_search(query,documents,3)
for res in semantic_search_res:
     print(f"Similarity score :{res[0]:.3f} | {res[1]}")
```
