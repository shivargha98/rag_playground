```python
from pinecone import Pinecone
from dotenv import load_dotenv
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
from pinecone import ServerlessSpec
from langchain_community.document_loaders import TextLoader
import os
load_dotenv()

embedding_model = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2') ##embedding model
```

```python
pinecone_api_key = os.environ['PINECONE_API_KEY']
pc = Pinecone(pinecone_api_key)

index_name = 'rag-index'
if not pc.has_index(index_name):
    pc.create_index(
        name = index_name,
        dimension=384,
        metric="cosine",
        spec = ServerlessSpec(cloud='aws',region='us-east-1')
    )
index = pc.Index(index_name)
```

```python
from langchain_pinecone import PineconeVectorStore
vectorstore = PineconeVectorStore(index,embedding_model)
```

```python
text_loader = TextLoader(
    file_path='data/transformer_architecture.txt',
    encoding = 'utf-8'
)
doc = text_loader.load()
vectorstore.add_documents(documents=doc)
```
